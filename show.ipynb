```
import re
import matplotlib.pyplot as plt
import numpy as np
```

```
def get_result_data(file):
  iters = []
  validation_accuracy = []
  ptn = r'(Iters: )(.+)(, validation_accuracy: )(.+)'
  with open(file, 'r') as f:
    line = f.readline()
    while line:
      t = re.search(ptn, line)
      if t != None:
        if float(t[2])>=140000:
          break
        iters.append(float(t[2]))
        validation_accuracy.append(float(t[4]))
      line = f.readline()
  f.close()
  return iters,validation_accuracy
fig = plt.figure()

show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_0)/test_lr_mini_saving_ckpt(1.1_0-1)(91.6).txt'
it1,ac1 = get_result_data(show_file)
show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_0)/test_lr_mini_saving_ckpt(1.1_0-2)(91.8).txt'
it2,ac2 = get_result_data(show_file)
it=it1+it2
ac=ac1+ac2
plt.plot(it, ac ,label='p:0',color='C0')
acm=max(ac)
plt.plot([it[0],it[0]], [acm,acm] ,label=acm,color='C0')

show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_2)/test_lr_mini_saving_ckpt(1.1_2-1)(93.0).txt'
it1,ac1 = get_result_data(show_file)
show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_2)/test_lr_mini_saving_ckpt(1.1_2-2)(93.4).txt'
it2,ac2 = get_result_data(show_file)
it=it1+it2
ac=ac1+ac2
plt.plot(it, ac ,label='p:2',color='C2')
acm=max(ac)
itm=max
plt.plot([it[0],it[0]], [acm,acm] ,label=acm,color='C2')

show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_2.5)/test_lr_mini_saving_ckpt(1.1_2.5-1)(94.6).txt'
it1,ac1 = get_result_data(show_file)
show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_2.5)/test_lr_mini_saving_ckpt(1.1_2.5-2)(94.6).txt'
it2,ac2 = get_result_data(show_file)
it=it1+it2
ac=ac1+ac2
plt.plot(it, ac ,label='p:2.5',color='C9')
acm=max(ac)
plt.plot([it[0],it[0]], [acm,acm] ,label=acm,color='C9')

show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_3_2)/test_lr_mini_saving_ckpt(1.1_3-1)(93.2).txt'
it1,ac1 = get_result_data(show_file)
show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_3_2)/test_lr_mini_saving_ckpt(1.1_3-2)(93.4).txt'
it2,ac2 = get_result_data(show_file)
it=it1+it2
ac=ac1+ac2
plt.plot(it, ac ,label='p:3',color='C3')
acm=max(ac)
plt.plot([it[0],it[0]], [acm,acm] ,label=acm,color='C3')

show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_4)/test_lr_mini_saving_ckpt(1.1_4-1)(91.8).txt'
it1,ac1 = get_result_data(show_file)
show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_4)/test_lr_mini_saving_ckpt(1.1_4-2)(92.4).txt'
it2,ac2 = get_result_data(show_file)
show_file='/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/test_lr_mini_saving_ckpt(1.1_4)/test_lr_mini_saving_ckpt(1.1_4-3)(92.0).txt'
it3,ac3 = get_result_data(show_file)
it=it1+it2+it3
ac=ac1+ac2+ac3
plt.plot(it, ac ,label='p:4',color='C4')
acm=max(ac)
plt.plot([it[0],it[0]], [acm,acm] ,label=acm,color='C4')

plt.legend(bbox_to_anchor=(1.05, 1))
plt.ylim(0.85,0.96)

fig.savefig('/content/gdrive/MyDrive/個別研究/車牌/LPRNet3/show_result.png', bbox_inches="tight",dpi=1000)
plt.show()
```
